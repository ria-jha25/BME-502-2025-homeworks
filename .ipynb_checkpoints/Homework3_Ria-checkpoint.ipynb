### (Bayes Theorem) Read the following example and complete the exercises 

Trolls Under the Bridge
1. Trolls(T) are bad
2. Gnomes(G) are benign
3. Every bridge has 5 creatures under it

Below are the probabilities of what creatures live under a bridge:
1. TTGGG: 25% (E1)
2. TGGGG: 25% (E2)
3. GGGGG: 50% (E3)

A knight is approaching a bridge and he wants to know his chance of crossing the bridge safely.

### 1.1 

#### What is the chance of the knight crossing the bridge safely?


```julia
println("The probability of the knight crossing the bridge safely is the same as the probability of all 5 creatures being Gnomes: 50%")
```

### 1.2

#### Before crossing a bridge, a knight captures a troll. “I now have an 75% chance of crossing safely,” he reasons, “since only the case 25% had TTGGG (E1) now have TGGG is still a threat.”

#### Do you agree with him? Why?

The knight is incorrect. His reasoning is flawed because he doesn't account for the effect that capturing a troll has on the probabilities of being in each scenario. If he has captured a troll, he he must be in either scenario E1 or E2, because E3 does not involve anay trolls. This would change the overall probability of crossing the bridge safely, as one of the original safe option has been eliminated. The new safe option would be E2 (GGGG). So, we have to consider its probability.

### 1.3

#### Now he wants to use Bayes' theorem to calculate his chance of crossing safely given a troll captured 

\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

#### Given the three events, E1, E2, and E3, which two of the three events does the knight need to consider after a troll was captured?

The knight has to consider E1 and E2 because those are the events involving trolls, and they would be affected by a troll's capture.

### 1.4

#### Use Bayes' theorem to calculate the probability of the two events in 1.3 (separately) given the troll was captured. 

Probability of one event given a troll captured: 

```julia
println("For E1 after troll capture = TGGG")

println("P(E1|T) = P(T|E1) * P(E1) / P(T)")
TgE1 = 2/5 # P(T|E1) 
println("P(T|E1) = $(round(TgE1, digits=4)) (2 trolls out of 5 creatures)")

E1 = 0.25 # P(E1)
println("P(E1) = $(round(E1, digits=4))")

E2 = 0.25 # P(E2)
println("P(E2) = $(round(E2, digits=4))")

TgE2 = 1/5 # P(T|E2)
println("P(T|E2) = $(round(TgE2, digits=4)) (1 troll out of 5 creatures)")

E3 = 0.5 # P(E3)
println("P(E3) = $(round(E3, digits=4))")
TgE3 = 0 # P(T|E3)
println("P(T|E3) = $(round(TgE3, digits=4))")

T = ((TgE1*E1) + (TgE2*E2) + (TgE3*E3))
println("P(T) = $(round(T, digits=4))")
E1gT = (TgE1 * E1) / T
println("P(T|E1) * P(E1) = $(round(TgE1*E1, digits=4))")
println("P(E1|T) = 0.1 / 0.15 = $(round(E1gT, digits=4))")
```

Probability of the other event given a troll captured: 

```julia
println("For E2 after troll capture = GGGG")


E2gT = (TgE2 * E2) / T
println("P(T|E2) * P(E2) = $(round(TgE2*E2, digits=4))")
println("P(E2|T) = 0.05 / 0.15 = $(round(E2gT, digits=4))")
```

### 1.5

#### What is the chance of the knight crossing safely?

```julia
println("P(E1|T) + P(E2|T) + P(E3|T) = 1")
E3gT = 1 - E1gT - E2gT
println("P(E3|T) = 1 - 0.6667 - 0.3333 = $(round(E3gT, digits=4))")
println("P(Safe) = P(E2|T) + P(E3|T) = $(round(E2gT, digits=4)) + $(round(E3gT, digits=4)) = $(round(E2gT, digits=4))")
```

# Question 2
### (Polynomial Fitting) Read the following example and complete the exercises 

Biological data always comes with errors that influence the intepretation of the data significantly. In this question, we will simulate data and see how noises affect the parameter estimation.

```julia
import Pkg
Pkg.activate(".")
Pkg.add(["Distributions", "Plots", "Polynomials", "Random"])
```

```julia
import Distributions as di
import Plots
import Polynomials
import Random
```

### 2.1

1. Generate the independent variable x ranging from -10 to 10 with an increment of 1
2. Generate the dependent variable y_true using the equation y_true = x^2 + 4x - 2


```julia
using Plots
using Random
using Polynomials: Polynomial, fit, coeffs
using Statistics
```

```julia
x = -10:10
collect(x)
```

```julia
y_true = (x.^2) .+ (4*x) .- (2.0)
```

### 2.2

1. Generate the Guassian noises with zero mean and a standard deviation of 10

2. Add the noises to y_true to obtain y_observed


```julia
n = length(y_true)
noise1 = rand(di.Normal(0, 10), n)
```

```julia
y_observed = y_true + noise1
```

### 2.3

#### Fit the data (x, y_observed) with a second order polynomial. 

1. Make a scatter plot of y_observed along x
2. Plot a line of y_true along x on the scatter plot
3. Plot a line of the fitted polynomial along x on the same plot

```julia
scatter(x, y_observed, label="Observed Data", color=:blue)
plot!(x, y_true, label="True Function", color=:red, linewidth=2)

p = fit(x, y_observed, 2)
plot!(x, p.(x), label="Fitted Polynomial", color=:green, linewidth=2)

xlabel!("x")
ylabel!("y")
```

### 2.4

1. Repeat 2.2 and 2.3 with a noise with a standard deviation of 40

```julia
noise2 = rand(di.Normal(0, 40), n)
y_observed2 = y_true + noise2

scatter(x, y_observed2, label="Observed Data", color=:blue)
plot!(x, y_true, label="True Function", color=:red, linewidth=2)

p2 = fit(x, y_observed2, 2)
plot!(x, p2.(x), label="Fitted Polynomial", color=:green, linewidth=2)

xlabel!("x")
ylabel!("y")
```

2. Compare the plot with the plot in 2.3. Is the fit closer to y_true or not, and why?

In the new plot with $\sigma = 40$, the fitted polynomial deviates more from the true function. The observed data show much larger scatter around the true curve.

The larger standard deviation ($\sigma = 40$) introduces more significant perturbations to the true values, causing the observed points to deviate further from the true function. With higher noise, the polynomial fit may also try to capture some of the random fluctuations, leading to a less accurate representation of the true function.

### 2.5

#### Repeat 2.2 and 2.3 with a tenth order polynomial fit


```julia
scatter(x, y_observed, label="Observed Data", color=:blue)
plot!(x, y_true, label="True Function", color=:red, linewidth=2)

p3 = fit(x, y_observed, 10)
plot!(x, p3.(x), label="Fitted Polynomial", color=:green, linewidth=2)

xlabel!("x")
ylabel!("y")
```

### 2.6

1. Create a new independent parameter x_fit ranging from -11 to 11 with an increment of 0.2
2. Plot the same three plots as in 2.3 again but apply the tenth order polynomial to x_fit instead of x


```julia
x_fit = -11:0.2:11
collect(x_fit)
```

```julia
scatter(x, y_observed, label="Observed Data", color=:blue),
plot!(x, y_true, label="True Function", color=:red, linewidth=2),
plot!(x_fit, p3.(x_fit), label="Fitted Polynomial", color=:green, linewidth=2)
xlabel!("x")
ylabel!("y")
```

3. What do you observe in the plot? What is this phenomenon called?

The 10th order polynomial fit passes through or close to all the observed data points. However, between the data points, especially at the edges of the x range, the polynomial fit shows larger oscillations and deviates significantly from the true function. The polynomial poorly fits values outside the range of the original data.

This is called overfitting, and these qualities are specifically charcteristics of Runge's phenomenon for high-order polynomial interpolation
